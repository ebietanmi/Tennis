```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier 
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
classifier = DecisionTreeClassifier()
score = accuracy_score
```

```python
dataset = pd.read_csv('../datasets/play_tennis.csv')
```

```python
# # dataset.columns
dataset['outlook'].replace(["Rain","Overcast","Sunny"], [1,2,3], inplace=True)
dataset['temp'].replace(["Cool","Mild","Hot"], [1,2,3], inplace=True)
dataset['wind'].replace(["Weak","Strong"], [0,1], inplace=True)
dataset['humidity'].replace(["Normal","High"],[0,1], inplace=True)
dataset['play'].replace(["No","Yes"],[0,1], inplace=True)
dataset.head(5)
```

```python
x = dataset.drop(columns=["play","day"])
y = dataset['play']
dataset
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.2)
```

```python
model = classifier.fit(x,y)
```

```python
# instruction for  use

# ['outlook'] = (["Rain","Overcast","Sunny"] = [1,2,3] respectively
# ['temp'] =(["Cool","Mild","Hot"]  = [1,2,3] respectively
# ['wind'] =(["Weak","Strong"]= [0,1] respectively
# ['humidity'] =(["Normal","High"] [0,1] respectively




predict = model.predict([[2,2,1,1]]) # = [outlook temp humidity wind ] respectively
if predict == True:
    print("Hello,you can play Tennis Today")
else:
    print("Hello, I am sorry you can not play Tennis Today")
    
```
